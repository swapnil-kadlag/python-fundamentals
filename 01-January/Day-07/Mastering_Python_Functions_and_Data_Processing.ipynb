

#Functional Foundations (`def`)


## 1. Definitions & Syntax

A **Function** is a named block of code that performs a specific task. It only runs when "called."

### **The Syntax Anatomy**

```
def function_name(parameter1, parameter2): # Parameters: Placeholders
    """Docstring: Explains the function's purpose"""
    result = parameter1 + parameter2
    return result # Sends the data back to the caller
```

---

## 2. Advanced Argument Concepts

1. **Default Arguments:** Setting a "backup" value (e.g., `def greet(name="User")`).
2. **Keyword Arguments:** Calling by name, not order (e.g., `greet(name="Abhishek")`).
3. **`*args`**: Collects unlimited positional arguments into a **Tuple**.
4. **`**kwargs`**: Collects unlimited keyword arguments into a **Dictionary**.

---

## 3. Deep-Dive Example: `Modular_Banking.py`

```
# Building a modular bank logic
def calculate_interest(balance, rate=0.05):
    """Calculates yearly interest safely."""
    return balance * rate

def process_transaction(current_balance, *amounts):
    """Handles multiple transactions at once using *args."""
    total_change = sum(amounts)
    new_balance = current_balance + total_change
    return new_balance

# Execution
my_bal = 1000
my_bal = process_transaction(my_bal, 500, -200, 50) # Added 500, removed 200, added 50
print(f"Final Balance: {my_bal} | Interest Earned: {calculate_interest(my_bal)}")
```

#  Functional Power (`Lambda`, `Map`, `Filter`)



## 1. The Lambda Function

A **Lambda** is a one-line, "anonymous" function. It is used for quick logic that you don't need to name.

* **Syntax:** `lambda arguments : expression`

## 2. High-Order Functions

These functions take **another function** as an input to process data quickly.

| Method | Purpose |
| --- | --- |
| **`map()`** | Applies a function to **every** item in a list. |
| **`filter()`** | Keeps only items that pass a **True/False** test. |
| **`reduce()`** | "Shrinks" a list into a **single value** (e.g., a total sum). |

---

## 3. Deep-Dive Example: `Data_Cleaner.py`

```
from functools import reduce

raw_data = [100, 250, 400, 800, 1200]

# 1. Lambda + Map: Apply 10% Tax to all
taxed_data = list(map(lambda x: x * 1.1, raw_data))

# 2. Lambda + Filter: Keep only values > 500
high_values = list(filter(lambda x: x > 500, taxed_data))

# 3. Lambda + Reduce: Total sum of high values
total_sum = reduce(lambda x, y: x + y, high_values)

print(f"High Value Sum: {total_sum}")
```

## 4. Interview Questions (Engineering Depth)

### **Q1: What is the "Mutable Default Argument" trap?**

**Answer:** Never use a mutable object (like a list `[]`) as a default parameter. Because the function definition happens once, that list is shared across **all** calls. If you append to it in one call, it stays there for the next call!

* **Solution:** Use `None` and initialize the list inside the function.

### **Q2: `is` vs `==` in Python?**

**Answer:** `==` checks for **Equality** (is the value the same?). `is` checks for **Identity** (are they stored at the exact same memory address?).

### **Q3: When should you use a Lambda instead of `def`?**

**Answer:** Use Lambda for simple, one-time transformations (inside `map` or `filter`). If the logic requires multiple lines, loops, or error handling, use `def` for readability and debugging.
