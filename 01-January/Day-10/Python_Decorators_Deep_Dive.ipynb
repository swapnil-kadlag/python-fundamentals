

# Python Decorators (`@decorator`)


**Focus:** Meta-programming, Function Wrapping, and the "Closure" Pattern.

---

## 1. Detailed Definition

### **What is a Decorator?**

A decorator is a function that takes **another function** as an argument, adds some extra functionality to it, and returns it—all without permanently modifying the source code of the original function.

### **The "Gift Wrap" Analogy:**

Imagine you have a **Toy (Your Function)**.

* A **Decorator** is like **Gift Wrapping**.
* The gift wrap adds a nice appearance and protection (Extra logic) but the toy inside remains exactly the same. When someone "opens" the gift, they still get the toy.

### **Why use them?**

1. **Logging:** Track when a function starts and ends.
2. **Timing:** Measure how many seconds a function takes to run.
3. **Authentication:** Check if a user is logged in before letting them run a sensitive function.
4. **Rate Limiting:** Prevent a function from being called too many times.

---

## 2. Syntax in Detail

The syntax uses the **`@`** symbol (Syntactic Sugar) placed above the function you want to "decorate."

### **The Manual Way (Under the hood):**

```
def my_decorator(func):
    def wrapper():
        print("Something before.")
        func()
        print("Something after.")
    return wrapper

def say_hello():
    print("Hello!")

# Manual connection
decorated_hello = my_decorator(say_hello)
decorated_hello()
```

### **The Pythonic Way (Using `@`):**

```
@my_decorator
def say_hello():
    print("Hello!")
```

## 3. Maximum Examples (Covering All Concepts)

### **Example A: The "Execution Timer" (Performance Testing)**

This is a classic real-world use case.

```
import time

def timer_decorator(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs) # Runs the actual function
        end_time = time.time()
        print(f"⏱️ Execution Time for '{func.__name__}': {end_time - start_time:.4f}s")
        return result
    return wrapper

@timer_decorator
def complex_calculation():
    time.sleep(1.5) # Simulating a heavy task
    return "Done!"

complex_calculation()
```

### **Example B: Handling Arguments (`*args`, `**kwargs`)**

To make a decorator work for **any** function, you must use `*args` and `**kwargs`.

```
def logger(func):
    def wrapper(*args, **kwargs):
        print(f"Calling {func.__name__} with {args}")
        return func(*args, **kwargs)
    return wrapper

@logger
def add_funds(amount, account_type="Savings"):
    return f"Added {amount} to {account_type}"

add_funds(500, account_type="Current")
```

### **Example C: Multiple Decorators (Stacking)**

You can apply more than one decorator to a single function. They execute from **top to bottom**.

```
@timer_decorator
@logger
def get_user_data():
    return {"id": 1, "name": "Abhishek"}
```

## 4. Interview Questions (Engineering Depth)

### **Q1: What are "First-Class Functions" and why are they needed for decorators?**

**Answer:** In Python, functions are first-class objects. This means they can be passed as arguments, returned from other functions, and assigned to variables. Decorators rely on this because they "pass" the original function into the decorator function.

### **Q2: What is the purpose of `functools.wraps`?**

**Answer:** When you decorate a function, the original function's metadata (like its name and docstring) is replaced by the wrapper's metadata. To prevent this and keep the original function's identity, we use `@functools.wraps(func)` inside the decorator.

### **Q3: Can decorators take their own arguments?**

**Answer:** Yes, but it requires a **third level** of nesting. You create a function that returns a decorator, which then returns the wrapper.
*Example:* `@repeat(times=3)`

### **Q4: What is a "Closure" in Python?**

**Answer:** A closure is a function object that remembers values in the enclosing scope even if they are not present in memory. Decorators are a practical implementation of closures.
