

#  (Supplement): Types of Inheritance

**Focus:** Architectural patterns and the Method Resolution Order (MRO).

---

## 1. Single Inheritance

**Definition:** A child class inherits from only **one** parent class.

* **Logic:** The simplest form. "A Savings Account **is a** Bank Account."
* **Example:**

```
class Device:
    def power_on(self):
        print("Device is now ON.")

class Phone(Device): # Single Inheritance
    def call(self):
        print("Making a call...")
```

---

## 2. Multiple Inheritance

**Definition:** A child class inherits from **more than one** parent class.

* **Logic:** A "SmartWatch" **is a** "Watch" AND **is a** "Computer."
* **Example:**

```
class Camera:
    def take_photo(self):
        print("Photo captured!")

class MusicPlayer:
    def play_music(self):
        print("Playing tracks...")

class SmartPhone(Camera, MusicPlayer): # Inheriting from two parents
    pass

# SmartPhone now has access to both parents' methods
```

---

## 3. Multilevel Inheritance

**Definition:** A child class inherits from a parent, which in turn inherits from another parent (Grandparent -> Parent -> Child).

* **Logic:** "Tesla Model 3" **is a** "Electric Car," which **is a** "Vehicle."
* **Example:**

```
class Vehicle:
    def move(self):
        print("Vehicle is moving.")

class Car(Vehicle):
    def steer(self):
        print("Car is steering.")

class ElectricCar(Car): # Child of Car, Grandchild of Vehicle
    def charge(self):
        print("Charging battery...")
```

---

## 4. Hierarchical Inheritance

**Definition:** Multiple child classes inherit from a **single** parent class.

* **Logic:** "Manager," "Developer," and "HR" all inherit from the single "Employee" class.
* **Example:**

```
class Staff:
    def sign_in(self):
        print("Staff member signed in.")

class Teacher(Staff):
    pass

class Principal(Staff):
    pass
# Both Teacher and Principal share the 'sign_in' logic from Staff
```

---

## 5. Hybrid Inheritance

**Definition:** A combination of two or more of the types mentioned above. Usually involves a mix of Multiple and Multilevel inheritance.

* **The Challenge:** This often creates the **"Diamond Problem."**
* **The Solution:** Python uses **MRO (Method Resolution Order)** and the C3 Linearization algorithm to decide which parent's method to call first.

---

## 🛠️ Advanced Example: Solving the Diamond Problem

```
class A:
    def speak(self):
        print("ID: A")

class B(A):
    def speak(self):
        print("ID: B")

class C(A):
    def speak(self):
        print("ID: C")

class D(B, C): # Hybrid Inheritance
    pass

d_obj = D()
d_obj.speak() # Output: ID: B (Because B comes first in the MRO)

# To see the order Python follows:
print(D.mro())
```

## 💬 Interview Questions (Inheritance Depth)

### **Q1: What is the "Diamond Problem" in Multiple Inheritance?**

**Answer:** It occurs when a class inherits from two parents that both inherit from the same grandparent. If the parents have overwritten a grandparent's method, the child doesn't know which one to use. Python solves this using **MRO**.

### **Q2: Why does Python prefer `super()` over calling the Parent name directly?**

**Answer:** `super()` is more flexible. If you change the name of the parent class, you don't have to update every `__init__` call. More importantly, in Multiple Inheritance, `super()` follows the **MRO**, ensuring every parent is initialized exactly once.

### **Q3: What is the `object` class in Python?**

**Answer:** In Python 3, every class you create implicitly inherits from the built-in `object` class. It is the "Great-Grandparent" of every object in Python.
