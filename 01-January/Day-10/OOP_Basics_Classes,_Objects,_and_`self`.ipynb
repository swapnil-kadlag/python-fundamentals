
#  Object-Oriented Programming (OOP) - Part 1


**Focus:** Classes, Objects, the `__init__` Method, and the `self` keyword.

---

## 1. Detailed Definition

### **What is a Class?**

A **Class** is a blueprint or a template for creating objects. It defines a set of **attributes** (data) and **methods** (behaviors) that the objects created from the class will have.

### **What is an Object?**

An **Object** is an **instance** of a class. If "Car" is the Class (the blueprint), then your "Tesla Model 3" is an Object (the actual physical instance).

### **Why use OOP?**

1. **Encapsulation:** Grouping related data and functions together.
2. **Organization:** Makes large codebases much easier to navigate.
3. **Reusability:** You can create 1,000 "User" objects from a single "User" class.

## 2. Syntax in Detail

### **The Anatomy of a Class**

```
class BankAccount:
    # The Constructor Method
    def __init__(self, owner, balance):
        self.owner = owner      # Attribute
        self.balance = balance  # Attribute

    # A Method (Function inside a class)
    def deposit(self, amount):
        self.balance += amount
        return f"New Balance: {self.balance}"
```

### **Understanding the Key Terms:**

* **`class` Keyword:** Used to define the blueprint.
* **`__init__` (Constructor):** A special method that runs automatically when you create a new object. It "initializes" the object's data.
* **`self` Keyword:** This represents the **specific instance** of the object. It allows the code to distinguish between "My Account" and "Your Account."

## 3. Maximum Examples (Covering All Concepts)

### **Example A: The Simple Entity (Employee Management)**

```
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def get_annual_tax(self):
        return self.salary * 0.15

# Creating Objects (Instantiation)
emp1 = Employee("Abhishek", 80000)
emp2 = Employee("Rahul", 65000)

print(f"{emp1.name}'s Tax: ${emp1.get_annual_tax()}")
```

### **Example B: Interactive Methods (The Smart Device)**

```
class SmartLight:
    def __init__(self, room_name):
        self.room = room_name
        self.is_on = False
        self.brightness = 50

    def toggle(self):
        self.is_on = not self.is_on
        status = "ON" if self.is_on else "OFF"
        print(f"{self.room} light is now {status}")

    def set_brightness(self, level):
        if 0 <= level <= 100:
            self.brightness = level
            print(f"Brightness set to {level}%")

# Usage
kitchen_light = SmartLight("Kitchen")
kitchen_light.toggle()
kitchen_light.set_brightness(85)
```

## 4. Interview Questions (Engineering Depth)

### **Q1: What is the difference between a Class Attribute and an Instance Attribute?**

**Answer:** An **Instance Attribute** (defined with `self`) is unique to each object (e.g., every account has a different balance). A **Class Attribute** is shared by all instances of the class (e.g., `bank_name = "Global Bank"` defined at the top of the class).

### **Q2: What happens if you forget to include `self` as the first argument in a method?**

**Answer:** Python will raise a `TypeError`. When you call `object.method()`, Python automatically passes the object itself as the first argument. If your method isn't expecting it, the program crashes.

### **Q3: What is the purpose of the `__str__` method?**

**Answer:** It is a "Magic Method" that defines what happens when you `print(object)`. Instead of showing a messy memory address like `<__main__.Car object at 0x...>`, it allows you to return a clean string like `"This is a Red Tesla"`.
