


# OOP Pillars — Inheritance & Encapsulation

**Focus:** Code Reusability (Inheritance) and Data Security (Encapsulation).

## 1. Inheritance: The "Is-A" Relationship

### **Detailed Definition**

Inheritance allows a **Child Class** to derive attributes and methods from a **Parent Class**. This prevents code duplication. If a `Manager` "is a" type of `Employee`, the `Manager` class should inherit from the `Employee` class.

### **The `super()` Function**

This is a built-in function that allows the Child class to call the `__init__` method (or any other method) of the Parent class. It ensures the Parent's logic is executed before the Child adds its specific features.

### **Syntax in Detail**

```
class Parent:
    def __init__(self, name):
        self.name = name

class Child(Parent): # Inheritance syntax
    def __init__(self, name, specific_attr):
        super().__init__(name) # Calls the Parent's constructor
        self.specific_attr = specific_attr
```

## 2. Encapsulation: The "Private" Shield

### **Detailed Definition**

Encapsulation is the practice of bundling data and methods inside a single unit (the class) and **restricting access** to certain components. This prevents external code from accidentally modifying internal states.

### **Naming Conventions in Python:**

1. **Public (`self.name`):** Accessible from anywhere.
2. **Protected (`self._name`):** A hint to other developers that this is "internal" (shouldn't be touched outside).
3. **Private (`self.__name`):** Strictly hidden. Python "mangles" the name to make it difficult to access from outside.

## 3. Maximum Examples (Covering All Concepts)

### **Example A: Inheritance in Banking (Savings vs. Current)**

```
class Account:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def display(self):
        print(f"Owner: {self.owner} | Balance: {self.balance}")

class SavingsAccount(Account):
    def __init__(self, owner, balance, interest_rate):
        super().__init__(owner, balance)
        self.interest_rate = interest_rate

    def apply_interest(self):
        self.balance += (self.balance * self.interest_rate)
        print(f"Interest applied! New balance: {self.balance}")

# Usage
my_savings = SavingsAccount("Abhishek", 5000, 0.04)
my_savings.display()
my_savings.apply_interest()
```

### **Example B: Encapsulation (Protecting the Password)**

```
class UserProfile:
    def __init__(self, username, password):
        self.username = username
        self.__password = password # Private Attribute

    def check_password(self, attempt):
        return attempt == self.__password

    def update_password(self, old_p, new_p):
        if self.check_password(old_p):
            self.__password = new_p
            print("Password updated successfully!")
        else:
            print("Current password incorrect.")

# Test
user = UserProfile("dev_abhi", "admin123")
print(user.username) # Works
# print(user.__password) # Throws AttributeError
```

## 4. Interview Questions (Engineering Depth)

### **Q1: What is Method Overriding?**

**Answer:** Method Overriding occurs when a Child class provides a specific implementation of a method that is already defined in its Parent class. The Child's version "overwrites" the Parent's version for that specific object.

### **Q2: Does Python support Multiple Inheritance?**

**Answer:** Yes. A class can inherit from more than one parent: `class Child(Parent1, Parent2)`. However, this can lead to the "Diamond Problem," which Python solves using **MRO (Method Resolution Order)**.

### **Q3: What is "Name Mangling" in Encapsulation?**

**Answer:** When you use double underscores (e.g., `__balance`), Python internally renames the variable to `_ClassName__variableName`. This is intended to prevent naming conflicts in subclasses, rather than being a true security feature.
