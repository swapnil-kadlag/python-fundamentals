

# Scope, Namespaces, and Modular Design

**Focus:** Global vs. Local Scope, the `nonlocal` keyword, and creating Python Modules.

---

## 1. Key Definitions

### **What is Scope?**

Scope refers to the region of a program where a specific variable is accessible. Python follows the **LEGB Rule**:

1. **L**ocal: Inside the current function.
2. **E**nclosing: Inside the "parent" function (for nested functions).
3. **G**lobal: At the top level of the script.
4. **B**uilt-in: Pre-defined Python names (like `print`, `len`).

### **What is a Module?**

A module is simply a `.py` file containing Python code. A **Package** is a folder containing multiple modules and a special `__init__.py` file.

---

## 2. Syntax & Advanced Keywords

### **The `global` Keyword**

Used inside a function to modify a variable that lives at the top level.

```
count = 0

def increment():
    global count # Tells Python to use the top-level 'count'
    count += 1
```

### **The `nonlocal` Keyword**

Used in nested functions to modify a variable in the outer (enclosing) function.

```
def outer():
    x = "local"
    def inner():
        nonlocal x # Modifies 'x' in outer()
        x = "modified"
    inner()
    print(x)
```

## 3. Deep-Dive Example: Building a Module

For your GitHub, you should create two files to show you understand how to **Import** code.

### **File 1: `banking_utils.py` (The Module)**

```
# This is a helper module
TAX_RATE = 0.05

def calculate_net(gross):
    return gross - (gross * TAX_RATE)
```

### **File 2: `main_app.py` (The Logic)**

```
# Importing from our custom module
from banking_utils import calculate_net, TAX_RATE

salary = 50000
net_pay = calculate_net(salary)

print(f"Tax Rate Applied: {TAX_RATE * 100}%")
print(f"Your Net Salary: ${net_pay}")
```

## 4. Interview Questions (Engineering Depth)

### **Q1: What is the "Namespace" in Python?**

**Answer:** A Namespace is a mapping from names (variables) to objects. Think of it as a dictionary where the keys are your variable names and the values are the data. Python creates a new namespace whenever a function is called.

### **Q2: What is the purpose of `if __name__ == "__main__":`?**

**Answer:** It prevents code from running automatically when a file is **imported** as a module. It ensures the script only executes its main logic when run directly.

### **Q3: Why is using `global` generally considered bad practice?**

**Answer:** It makes code hard to debug. If a variable can be changed by any function anywhere, it’s difficult to track where a bug was introduced. It's better to pass variables as arguments and `return` the results.
