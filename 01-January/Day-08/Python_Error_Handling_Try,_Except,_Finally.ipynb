

# Exception Handling (`try`, `except`, `finally`)

**Focus:** Error Types, Graceful Degradation, and Custom Exceptions.

---

## 1. Key Definitions

### **What is an Exception?**

An Exception is an error that occurs **during execution**. Even if your syntax is perfect, an exception can happen (e.g., `10 / 0`). If not "caught," the program crashes with a "Traceback."

### **The "Safety Net" Analogy:**

Think of your code like a **Trapeze Artist**:

* **`try`**: The performance (The risky code).
* **`except`**: The safety net (Catches the fall).
* **`else`**: The applause (Runs only if there was no fall).
* **`finally`**: Cleaning up the stage (Runs no matter what, even if they fell).

---

## 2. Syntax & Common Error Types

### **Common Exceptions to Know:**

| Exception | Reason |
| --- | --- |
| **`ValueError`** | Right type, but wrong value (e.g., `int("abc")`). |
| **`TypeError`** | Operation on incompatible types (e.g., `5 + "10"`). |
| **`ZeroDivisionError`** | Dividing by zero. |
| **`FileNotFoundError`** | Trying to open a file that doesn't exist. |
| **`KeyError`** | Looking for a missing key in a dictionary. |

## 3. Deep-Dive Example: `safe_vault_access.py`

*This script demonstrates how to handle multiple errors and ensure system resources are closed.*

```
#  Robust Error Handling
# ----------------------------

def process_withdrawal(balance, amount):
    try:
        # 1. Potential ValueError if amount isn't a number
        # 2. Potential ZeroDivisionError or custom logic errors
        if amount > balance:
            raise ValueError("Insufficient Funds!") # Manually 'raising' an error

        new_balance = balance - amount
        print(f"Transaction successful. Remaining: ${new_balance}")

    except ValueError as e:
        print(f"❌ Transaction Failed: {e}")

    except TypeError:
        print("❌ Error: Please enter a valid numerical amount.")

    except Exception as e:
        # The 'Catch-All' (Use sparingly!)
        print(f"❌ An unexpected error occurred: {e}")

    else:
        # Runs ONLY if no exception occurred
        print("✅ Email notification sent to user.")

    finally:
        # Always runs (Great for closing database connections or files)
        print("🔒 Vault connection closed.")

# --- Testing ---
process_withdrawal(1000, 1500) # Triggers ValueError
process_withdrawal(1000, "hundred") # Triggers TypeError
```

## 4. Interview Questions (Engineering Depth)

### **Q1: What is the difference between `SyntaxError` and an `Exception`?**

**Answer:** A `SyntaxError` is caught by the Python interpreter **before** the code runs (like a grammar mistake). An `Exception` happens **while** the code is running (like running out of gas while driving).

### **Q2: Why shouldn't you use a "Bare Except" (`except:`)?**

**Answer:** Using `except:` without a specific error type catches *everything*, including system exits and `Ctrl+C` interrupts. This makes debugging nearly impossible because you don't know *what* failed. Always catch specific errors like `except ValueError:`.

### **Q3: When would you use the `finally` block?**

**Answer:** For "cleanup" actions that must happen regardless of success or failure—such as closing a file, disconnecting from a database, or releasing a network socket.
