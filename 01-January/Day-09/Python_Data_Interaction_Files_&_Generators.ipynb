

# File I/O & Persistent Data

**Focus:** Reading/Writing Files, Context Managers, and File Modes.

### 1. Key Definitions

* **File I/O:** The process of "Input" (Reading from a disk) and "Output" (Writing to a disk).
* **Context Manager (`with`):** A Pythonic way to handle files that automatically closes the file, even if an error occurs.

### 2. File Modes

| Mode | Description |
| --- | --- |
| **`'r'`** | **Read** (Default). Error if file doesn't exist. |
| **`'w'`** | **Write**. Creates a new file or overwrites existing. |
| **`'a'`** | **Append**. Adds data to the end of the file. |
| **`'r+'`** | **Read and Write**. |

### 3. Deep-Dive Example: `log_manager.py`

```
# Handling Real Data
# -------------------------

# Writing to a file using the 'with' statement (Context Manager)
with open("daily_report.txt", "w") as file:
    file.write("January 10: System initialized.\n")
    file.write("All services are running normally.")

# Reading from the file
try:
    with open("daily_report.txt", "r") as file:
        content = file.readlines() # Returns a list of lines
        for line in content:
            print(f"Log Entry: {line.strip()}")
except FileNotFoundError:
    print("Error: The requested file was not found.")
```

# Iterators & Generators (Memory Optimization)

**Focus:** Lazy Evaluation, the `yield` keyword, and Handling Large Datasets.

### 1. Key Definitions

* **Iterator:** An object that allows you to traverse a collection one by one using `next()`.
* **Generator:** A special function that returns an "iterator" object. It doesn't store all values in memory; it generates them **on the fly**.

### 2. The `yield` Keyword

Unlike `return`, which kills a function, `yield` "pauses" the function, returns a value, and waits to be called again. This is called **Lazy Evaluation**.

### 3. Deep-Dive Example: `big_data_processor.py`

```
# Processing 1 Million records without crashing RAM
# --------------------------------------------------------

def transaction_generator(limit):
    """A generator that simulates millions of transactions."""
    count = 1
    while count <= limit:
        yield f"TransactionID_{count}" # Pauses here
        count += 1

# Using the generator
gen = transaction_generator(1000000)

print(next(gen)) # TransactionID_1
print(next(gen)) # TransactionID_2

# Iterating without loading 1 million strings into memory
for tx in transaction_generator(5):
    print(f"Processing: {tx}")
```

## 4. Interview Questions (Engineering Depth)

### **Q1: Why is the `with` statement preferred over `file.close()`?**

**Answer:** If an error happens between `open()` and `close()`, the file remains open, leading to memory leaks and file corruption. The `with` statement (Context Manager) guarantees the file is closed no matter what.

### **Q2: What is the difference between `read()`, `readline()`, and `readlines()`?**

**Answer:** `read()` gets the entire file as one string. `readline()` gets just the next line. `readlines()` gets the entire file as a **list of strings**. For large files, `readline()` or a `for` loop over the file object is safest for memory.

### **Q3: When should you use a Generator instead of a List?**

**Answer:** Use a **Generator** when the dataset is extremely large (e.g., reading a 10GB CSV file). A List stores all items in RAM, which will crash your system. A Generator only holds the **current** item in RAM.
