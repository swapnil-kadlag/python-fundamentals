## 1. Definition

A **Tuple** is a collection which is **ordered** and **unchangeable (immutable)**. Once a tuple is created, you cannot add, remove, or modify its elements.

### Why use a Tuple?

* **Data Integrity:** It acts as a "constant" list. Use it for data that should never change (e.g., RGB color codes, GPS coordinates).
* **Performance:** Because they are fixed in size, they are faster to process than lists.
* **Dictionary Keys:** Unlike lists, tuples can be used as keys in a dictionary because they are "hashable."

---

## 2. Basic Examples

### A. Creating Tuples

```python
# Standard Tuple
coordinates = (18.5204, 73.8567) # Pune Latitude/Longitude

# The "Singleton" Trap (Very Important!)
# To create a tuple with ONE item, you MUST include a trailing comma.
wrong_tuple = ("Apple")  # This is a String
right_tuple = ("Apple",) # This is a Tuple

# Tuple Packing and Unpacking
person = ("Abhishek", 25, "AI Engineer")
name, age, role = person # Unpacking into variables
```

### B. Accessing Data

Tuples use **zero-based indexing**, exactly like lists.

```python
fruits = ("Mango", "Banana", "Cherry")
print(fruits[0])  # Output: Mango
print(fruits[-1]) # Output: Cherry (Last item)
```

---

## 3. Tuple Methods

Because tuples cannot be modified, they only have **two** built-in methods. They do not have `.append()`, `.pop()`, or `.sort()`.

### I. The `.count()` Method

Returns the number of times a specified value occurs in a tuple.

```python
numbers = (1, 2, 3, 1, 4, 1, 5)
print(numbers.count(1)) # Output: 3
```

### II. The `.index()` Method

Searches the tuple for a specified value and returns the **first** position where it was found.

```python
vowels = ('a', 'e', 'i', 'o', 'u')
print(vowels.index('i')) # Output: 2
```

> **Note:** If you search for a value that doesn't exist using `.index()`, Python will raise a `ValueError`.

```python
print("--- SECTION 3: MEMORY MANAGEMENT ---")

# The Scenario: A nested list (List inside a List)
original_list = [["Project A", "Active"], ["Project B", "Pending"]]

# 3.1 Shallow Copy
# This creates a new outer list but keeps references to the inner lists.
shallow_copy = original_list.copy()
shallow_copy[0][1] = "COMPLETED"

print("SHALLOW COPY BEHAVIOR:")
print(f"Original: {original_list}") # Original CHANGED because the inner list is shared!
print(f"Shallow:  {shallow_copy}")


import copy
# 3.2 Deep Copy
# This creates a completely independent clone of every object inside.
deep_copy = copy.deepcopy(original_list)
deep_copy[1][1] = "CANCELLED"

print("\nDEEP COPY BEHAVIOR:")
print(f"Original: {original_list}") # Original remains "Pending"
print(f"Deep:     {deep_copy}")
```
