## 1. The `for` Loop (Definite Iteration)

A `for` loop is used to iterate over a sequence (list, tuple, dictionary, set, or string). It runs a specific number of times based on the length of the collection.

**Syntax:**

for item in sequence:

    # Code to execute

### Example: Iterating with `range()`

The `range(start, stop, step)` function is the most common way to generate a sequence of numbers.

```
# Print even numbers from 2 to 10
for i in range(2, 11, 2):
    print(i) # Output: 2, 4, 6, 8, 10
```

---

## 2. The `while` Loop (Indefinite Iteration)

A `while` loop runs as long as a condition remains **True**. It is used when you don't know exactly how many times you need to repeat (e.g., waiting for user input).

**Syntax:**

```
while condition:
    # Code to execute
    # Must include a way to change the condition to avoid infinite loops!
```

### Example: A Simple Countdown

```
countdown = 5
while countdown > 0:
    print(countdown)
    countdown -= 1  # Decrementing to eventually break the loop
print("Blast off!")
```

---

## 3. Loop Control: `break`, `continue`, and `pass`

These keywords give you "fine-grained" control over your loops.

* **`break`**: Stops the loop entirely and moves to the next section of code.
* **`continue`**: Skips the current iteration and jumps to the next one.
* **`pass`**: A "null" statement. It does nothing but acts as a placeholder to prevent syntax errors.

### Example: Searching for a specific item

```
numbers = [1, 5, 8, 10, 3, 12]

for n in numbers:
    if n % 2 == 0:
        print(f"First even number found: {n}")
        break  # Exit loop immediately after finding the first even number
```

---

## 4.  Iterator Helpers: `enumerate()` and `zip()`

In interviews, using these makes you look like a pro.

* **`enumerate(list)`**: Gives you both the **index** and the **value** at the same time.
* **`zip(list1, list2)`**: Pairs elements from two lists together like a zipper.

### Example: Combining Data

```
names = ["Abhishek", "Rahul"]
scores = [95, 88]

for index, (name, score) in enumerate(zip(names, scores)):
    print(f"{index}: {name} scored {score}")
```

## ❓ Interview Questions: Loops

**Q1: What is an "Infinite Loop" and how do you stop it?**
**Answer:** An infinite loop occurs when the `while` condition never becomes `False`. In most terminals, you stop it using **Ctrl + C**.

**Q2: What is the difference between `range(5)` and `list(range(5))`?**
**Answer:** `range(5)` is a **generator object** (lazy evaluation). It doesn't store all numbers in memory at once, which is efficient for large ranges. `list(range(5))` creates an actual list in memory `[0, 1, 2, 3, 4]`.

**Q3: When would you use a `while` loop instead of a `for` loop?**
**Answer:** Use `for` when you know the size of the data (Iterating a list). Use `while` when the end depends on an external factor (e.g., "Loop until the user clicks 'Exit'").
