## 1. Key Definitions

### **What is a List?**

A Python **List** is an ordered, mutable (changeable) collection of items. Unlike a simple variable that holds one value (e.g., `x = 5`), a list can hold thousands of items of different data types (integers, strings, or even other lists).

### **Key Characteristics:**

1. **Ordered:** The items have a specific index (starting at 0).
2. **Mutable:** You can add, remove, or change items after the list is created.
3. **Dynamic:** Lists can grow or shrink in size as needed.
4. **Heterogeneous:** A single list can contain a mix of `int`, `str`, and `float`.

---

## 2. Essential List Methods

In Python, we use "Methods" (functions attached to the list) to manipulate data.

| Method | Action | Example |
| --- | --- | --- |
| **`.append(x)`** | Adds item `x` to the **end** of the list. | `prices.append(99.9)` |
| **`.insert(i, x)`** | Adds item `x` at a **specific index** `i`. | `names.insert(0, "First")` |
| **`.pop()`** | Removes and **returns** the last item. | `last_item = my_list.pop()` |
| **`.remove(x)`** | Finds and removes the **first occurrence** of `x`. | `users.remove("Abhishek")` |
| **`.sort()`** | Arranges items in **ascending order** (in-place). | `numbers.sort()` |
| **`.reverse()`** | Flips the order of the list. | `numbers.reverse()` |

## 3. Detailed Code Example: `bank_vault.py`

This script demonstrates a real-world use case for lists: managing a queue of bank transactions.

```
# January 7: The Bank Vault Manager

# 1. Initialize an empty list
vault_transactions = []

# 2. Using .append() to add data
vault_transactions.append(5000)
vault_transactions.append(1200)
vault_transactions.append(3500)

# 3. Using .insert() to put a priority transaction at the start
vault_transactions.insert(0, 10000)

# 4. Accessing data via Indexing
# Indexing starts at 0!
print(f"First Transaction: {vault_transactions[0]}")
print(f"Last Transaction: {vault_transactions[-1]}") # -1 is a shortcut for 'last'

# 5. Slicing (Getting a sub-portion of the list)
top_two = vault_transactions[0:2]
print(f"Top 2 Transactions: {top_two}")

# 6. Safety Check: If-In logic
search_amt = 1200
if search_amt in vault_transactions:
    print(f"Verified: {search_amt} exists in vault.")

# 7. Cleaning up
vault_transactions.pop() # Removes the last entry
print(f"Final Count: {len(vault_transactions)}")
```

## 4. Interview Questions (The "Technical Deep Dive")

### **Q1: What is the difference between `.append()` and `.extend()`?**

**Answer:** `.append()` adds a single object to the end of the list. If you append a list to another list, you get a "nested" list. `.extend()` takes an iterable (like another list) and adds each of its elements individually to the end.

### **Q2: Why is Indexing in Python 0-based?**

**Answer:** It relates to **Memory Offsets**. The index represents the "distance" from the start of the memory block. The first element is at the very start (Distance = 0).

### **Q3: What is "Mutability" and why does it matter?**

**Answer:** Mutability means the object can be changed without creating a new one in memory. Lists are **mutable**, which makes them efficient for adding/removing data. Strings are **immutable**; every time you "change" a string, Python actually creates a brand new string object in the background.

### **Q4: How do you copy a list safely?**

**Answer:** If you write `list_b = list_a`, both variables point to the **same memory address**. Changing `list_b` will change `list_a`. To create a real copy, use `list_b = list_a.copy()` or `list_b = list_a[:]`.
