# Complete Guide to Abstraction in Python

**Abstraction** is a core concept of Object-Oriented Programming (OOP). It means **hiding the complex implementation details** and showing only the essential features of the object to the user.

Think of a car: You know how to use the steering wheel and the accelerator pedal (the interface). You don't need to know exactly how the engine mixes fuel and air or how the transmission gears shift (the hidden implementation).

In Python, we achieve abstraction using **Abstract Base Classes (ABC)** from the `abc` module.

---
## 1. What is an Abstract Class?
An abstract class is a blueprint for other classes. It allows you to declare methods that must be created within any child classes built from the abstract class.

**Key Rules of Abstraction:**
1. You **cannot create an object (instantiate)** directly from an abstract class.
2. To make a class abstract, it must inherit from `ABC`.
3. It uses the `@abstractmethod` decorator to define methods that *must* be implemented by its children.

```python
from abc import ABC, abstractmethod

# 1. Create an Abstract Base Class
class Computer(ABC):
    
    # 2. Define an abstract method
    @abstractmethod
    def process(self):
        # This method is empty. It's just a rule saying "All computers MUST have a process method."
        pass

    # Note: Abstract classes CAN have normal (concrete) methods too!
    def boot_up(self):
        print("Booting up the computer system...")


# Let's try to create a Computer directly. (Spoiler: This will crash!)
try:
    my_pc = Computer()
except TypeError as e:
    print(f"Error: {e}")
```

## 2. Implementing the Abstract Class
To use our `Computer` blueprint, we must create a child class that provides the actual code for the `process()` method.

```python
class Laptop(Computer):
    # Overriding the abstract method. If we don't do this, Laptop will ALSO be abstract!
    def process(self):
        print("Laptop processing data... Beep boop.")

class Desktop(Computer):
    def process(self):
        print("Desktop processing complex tasks at high speeds!")

# Now we can instantiate them
my_laptop = Laptop()
my_desktop = Desktop()

# We can use both the concrete method from the parent AND the overridden abstract method
my_laptop.boot_up()
my_laptop.process()

print("-----")

my_desktop.boot_up()
my_desktop.process()
```

## 3. Real-World Example: A Payment System
Imagine you are building an e-commerce store. You want to support Credit Cards, PayPal, and Bitcoin. No matter *how* they process the payment internally, they must all have a `pay()` method.

```python
class PaymentGateway(ABC):
    @abstractmethod
    def pay(self, amount):
        pass

class CreditCardPayment(PaymentGateway):
    def pay(self, amount):
        print(f"Processing a Credit Card payment of ${amount}")

class PayPalPayment(PaymentGateway):
    def pay(self, amount):
        print(f"Redirecting to PayPal to process ${amount}")

class BitcoinPayment(PaymentGateway):
    def pay(self, amount):
        print(f"Verifying blockchain transaction for ${amount} equivalent in BTC")

# Using the abstraction
def checkout(payment_method, amount):
    # The checkout system DOES NOT CARE how the payment is processed.
    # It only knows that `payment_method` has a `pay()` function.
    payment_method.pay(amount)

checkout(CreditCardPayment(), 100)
checkout(PayPalPayment(), 50)
checkout(BitcoinPayment(), 200)
```

---
## 4. Why Use Abstraction?

1. **Enforces a Standard Interface:** If you collaborate with a team, you can write the `Abstract Class` as the architect. You assign the child classes to your team. You *guarantee* that they will name their method `process()` instead of `run()`, `execute()`, or `do_it()`, because Python forces them to override `process()`.
2. **Hides Complexity:** The user of the class only needs to know *what* the object does, not *how* it does it.
3. **Prevents Instantiation of Incomplete Classes:** You can't accidentally create an object of a class that is meant to only be a blueprint.

## 🎓 Top Interview Questions on Python Abstraction

### Q1: What is Abstraction in OOP?
**Answer:** Abstraction is the concept of hiding the internal, complex implementation details of a system and exposing only the essential functionalities to the user. In Python, this is achieved using abstract base classes using the `abc` module.

### Q2: What is an Abstract Base Class (ABC)?
**Answer:** An ABC is a class that defines a common interface for its subclasses. It cannot be instantiated on its own. It serves as a blueprint, forcing its subclasses to implement specific methods using the `@abstractmethod` decorator.

### Q3: How do you create an abstract class in Python?
**Answer:**
1. Import `ABC` and `abstractmethod` from the `abc` module.
2. The class must inherit from `ABC`.
3. At least one method must be decorated with `@abstractmethod`.

### Q4: Can an abstract class have regular (concrete) methods?
**Answer:** Yes. Abstract classes can have both abstract methods (with no implementation) and concrete methods (with full implementation). Subclasses inherit the concrete methods and can use them directly.

### Q5: What happens if a subclass does not implement all the abstract methods of its parent class?
**Answer:** The subclass itself will also be considered an abstract class. If you try to instantiate it, Python will throw a `TypeError`.

### Q6: What is the difference between Abstraction and Encapsulation?
**Answer:** 
- **Abstraction** is about *hiding complexity* at the design level. It focuses on *what* an object does (e.g., "You can `drive()` this car").
- **Encapsulation** is about *hiding data* at the implementation level. It restricts direct access to variables to protect the internal state of an object (e.g., "You cannot directly change `car.__engine_temperature`").
