#  Polymorphism in Python

Polymorphism comes from the Greek words *poly* (many) and *morph* (forms). It means the same function name or operator can be used for different types.

---
## 1. Built-in Polymorphism
In Python, the same built-in function or operator can show different behaviors depending on the object it acts upon.

```python
# Example 1: The len() function
print("Length of string 'Hello':", len("Hello"))
print("Length of list [1, 2, 3]:", len([1, 2, 3]))
print("Length of dict {'a': 1, 'b': 2}:", len({'a': 1, 'b': 2}))

# Example 2: The + operator
print("Integer addition:", 10 + 20)
print("String concatenation:", "Hello " + "World")
```

## 2. Polymorphism with Class Methods
Different classes can have methods with the same name. We can call these methods without knowing the exact object type.

```python
class Car:
    def start(self):
        return "Vroom! Car started."

class Bike:
    def start(self):
        return "Ring ring! Bike started."

# A common interface to start any vehicle
def start_vehicle(vehicle):
    print(vehicle.start())

my_car = Car()
my_bike = Bike()

start_vehicle(my_car)
start_vehicle(my_bike)
```

## 3. Polymorphism with Inheritance (Method Overriding)
A child class can inherit methods from a parent class but define its own specific behavior for a method with the same name.

```python
import math

class Shape:
    def area(self):
        pass # Placeholder for child classes

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return math.pi * (self.radius ** 2)

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
        
    def area(self):
        return self.width * self.height

shapes = [Circle(5), Rectangle(4, 6)]

for shape in shapes:
    print(f"The area of the {type(shape).__name__} is: {shape.area():.2f}")
```

## 4. Duck Typing (Polymorphism without strict Inheritance)
In Python, "If it walks like a duck and quacks like a duck, it must be a duck." 
We don't care about the object's true class, we only care if it implements the required methods.

```python
class Duck:
    def sound(self):
        return "Quack!"

class Dog:
    def sound(self):
        return "Woof!"

class ToyDuck:
    def sound(self):
        return "Squeak!"

def make_sound(entity):
    # It doesn't check if the entity IS-A Duck, just if it has a sound() method
    print(entity.sound())

make_sound(Duck())
make_sound(Dog())
make_sound(ToyDuck())
```

## 5. Operator Overloading (Magic Methods)
We can change the behavior of Python's built-in operators (like `+`, `-`, `*`, `==`) for our own custom objects.

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    # Overloading the '+' operator
    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)
        
    def __str__(self):
        return f"({self.x}, {self.y})"

p1 = Point(1, 2)
p2 = Point(3, 4)
p3 = p1 + p2  # Uses the __add__ method!

print(f"p1: {p1}")
print(f"p2: {p2}")
print(f"p1 + p2 = {p3}")
```

## 🎓 Top Interview Questions on Python Polymorphism

### Q1: What is Polymorphism in Python?
**Answer:** Polymorphism means "many forms." In Python, it allows different objects to use the same method names, and built-in operators or functions to behave differently depending on the operand/argument types (e.g., `len()` returning the length of a list or a string).

### Q2: Does Python support Method Overloading (like Java or C++)?
**Answer:** No, Python does not support strict method overloading (having multiple methods with the exact same name but different parameters in the same class). If you define a method twice, the latest definition overwrites the previous one. 
*Workaround:* We use default arguments, variable-length arguments (`*args`, `**kwargs`), or type checking inside a single method to achieve the same effect.

### Q3: What is the difference between Method Overriding and Method Overloading?
**Answer:** 
- **Overriding:** Occurs in inheritance. A child class provides a new implementation for a method defined in its parent class (Runtime Polymorphism).
- **Overloading:** Having multiple methods with the same name but different signatures in the same class. (Compile-time Polymorphism). Python does not natively support this.

### Q4: Explain "Duck Typing" in the context of Polymorphism.
**Answer:** Duck Typing is a concept where the type or the class of an object is less important than the methods it defines. When you call `obj.do_something()`, Python doesn't care what type `obj` is, as long as it has a `do_something()` method. This is a core aspect of Python's dynamic polymorphism.

### Q5: How do you overload an operator in Python?
**Answer:** By defining special "dunder" (double underscore) methods in your class. For example, implementing `__add__(self, other)` allows you to use the `+` operator on your objects, and `__eq__(self, other)` allows you to use the `==` operator.

### Q6: Can we achieve Polymorphism without Inheritance?
**Answer:** Yes, via Duck Typing. As long as different classes implement the same method names, we can iterate over them and call the methods uniformly, without them sharing a common parent class.

### Q7: What is the difference between Inheritance and Polymorphism?

**Answer:** Inheritance is about **code reuse** (taking features from a parent). Polymorphism is about **flexibility** (allowing different objects to be treated as instances of the same parent while maintaining their unique behaviors).
