## 1. The If-Elif-Else

Python evaluates conditions from top to bottom. As soon as one condition is **True**, it executes that block and **skips** the rest of the ladder.

###  Example: A Grade & Reward System

```
# Multiple conditions using logical operators
score = 85
attendance = 92

if score >= 90 and attendance >= 90:
    result = "Distinction with Honors"
elif score >= 75 or attendance >= 95:
    result = "First Class"
elif not (score < 40):
    result = "Passed"
else:
    result = "Failed"

print(f"Final Status: {result}")
```

## 2. Ternary Operators (Conditional Expressions)

In Python, we prefer "One-Liners" for simple assignments. It makes the code "Pythonic."

**Syntax:** `X if Condition else Y`

```
# Traditional Way (4 lines)
age = 17
if age >= 18:
    status = "Adult"
else:
    status = "Minor"

print(status)

# Ternary Way (1 line)
status = "Adult" if age >= 18 else "Minor"

print(status)
```

## ❓ Interview Questions: Control Flow

### Q1: What is "Short-Circuiting" in Python logic?

**Answer:** Python is lazy (efficient). In an `and` statement, if the first part is `False`, Python **does not check** the second part because the whole thing must be false. In an `or` statement, if the first part is `True`, it skips the rest.

* *Interview Tip:* Use this to prevent errors, e.g., `if count != 0 and (total / count) > 10:`. The first check prevents a "Division by Zero" error.

### Q2: Difference between `if-if-if` and `if-elif-else`?

**Answer:** * **`if-if-if`**: Python checks **every** condition, even if the first one was true. (Slower/Redundant).

* **`if-elif-else`**: Python stops as soon as it finds a match. (Faster/Cleaner).

### Q3: What is the `_` in a `match-case` statement?

**Answer:** It is the **Wildcard**. It acts as the "else" or "default" case. It catches anything that wasn't matched by the specific cases above it.

```
# Interview Challenge: The FizzBuzz Logic
# Task: If num div by 3 (Fizz), by 5 (Buzz), by both (FizzBuzz)

def fizz_buzz(n):
    if n % 3 == 0 and n % 5 == 0:
        return "FizzBuzz"
    elif n % 3 == 0:
        return "Fizz"
    elif n % 5 == 0:
        return "Buzz"
    else:
        return str(n)

# Testing the logic with a Ternary operator
num = 15
print(f"Result for {num}: {fizz_buzz(num)}")

# Nested Logic Example
user_authenticated = True
admin_rights = False

if user_authenticated:
    print("Welcome, User.")
    if admin_rights:
        print("Admin Dashboard Enabled.")
    else:
        print("Standard Dashboard Enabled.")
else:
    print("Please Log In.")
```
