
# Data Mapping with Dictionaries (`dict`)


**Focus:** Key-Value Pairs, Uniqueness, Fast Lookups, and Dictionary Methods.

## 1. Key Definitions

### **What is a Dictionary?**

A Python **Dictionary** is an unordered (in older versions) or insertion-ordered (Python 3.7+) collection of **Key-Value pairs**. Unlike a list where you find an item by its position (0, 1, 2), in a dictionary, you find a **Value** by using a unique **Key**.

### **The "Phonebook" Analogy:**

* **Key:** The Name (Must be unique).
* **Value:** The Phone Number (Can be shared by multiple names).

### **Key Characteristics:**

1. **Unique Keys:** You cannot have two identical keys. If you add a duplicate, the old value is overwritten.
2. **Immutable Keys:** Keys must be "Hashable" (Strings, Integers, or Tuples). You cannot use a List as a key.
3. **Fast Lookups:** Finding a value by its key is nearly instantaneous, regardless of how large the dictionary is ($O(1)$ complexity).

## 2. Syntax & Common Methods

### **Basic Syntax:**

```
# {Key: Value}
user_data = {"id": 101, "name": "Abhishek", "role": "Developer"}
```

### **Essential Methods:**

| Method | Action | Example |
| --- | --- | --- |
| **`.get(key, default)`** | Safely gets a value. Returns `default` if key is missing (Prevents `KeyError`). | `data.get("age", 0)` |
| **`.keys()`** | Returns a list-like view of all **keys**. | `user.keys()` |
| **`.values()`** | Returns a list-like view of all **values**. | `user.values()` |
| **`.items()`** | Returns **tuples** of (key, value) pairs. | `user.items()` |
| **`.update({k: v})`** | Merges another dictionary into the current one. | `user.update({"city": "Pune"})` |
| **`.pop(key)`** | Removes the key and returns its value. | `val = user.pop("role")` |

## 3. Detailed Code Example: `bank_database.py`

This script demonstrates how to manage multiple user accounts using a dictionary.

```
# Dictionary Implementation for Banking

# 1. Creating a nested dictionary (Dictionary inside a Dictionary)
accounts = {
    1001: {"name": "Abhishek", "balance": 5000, "type": "Savings"},
    1002: {"name": "Rahul", "balance": 1200, "type": "Current"}
}

# 2. Accessing Data
user_id = 1001
print(f"User: {accounts[user_id]['name']} | Balance: {accounts[user_id]['balance']}")

# 3. Safe Access using .get() (Best Practice)
# This prevents the program from crashing if the ID doesn't exist
unknown_user = accounts.get(9999, "User Not Found")
print(f"Search Result: {unknown_user}")

# 4. Modifying Data
accounts[1001]["balance"] += 500  # Deposit
print(f"New Balance for 1001: {accounts[1001]['balance']}")

# 5. Iterating through the database
print("\n--- Full Account Audit ---")
for acc_id, info in accounts.items():
    print(f"ID: {acc_id} | Name: {info['name']} | Status: Active")
```

## 4. Interview Questions (The "Pro" Level)

### **Q1: What happens if you try to access a key that doesn't exist?**

**Answer:** Python raises a **`KeyError`**. To avoid this, you should use the **`.get()`** method, which returns `None` (or a default value you choose) instead of crashing the program.

### **Q2: Why can't a List be used as a Dictionary Key?**

**Answer:** Keys must be **Hashable** and **Immutable**. Since a list is mutable (it can change), its "hash value" would change, making it impossible for Python to find it again in memory. Strings and Tuples are preferred.

### **Q3: How do Dictionaries differ from Lists in terms of performance?**

**Answer:** * **Lists:** Finding an element takes $O(n)$ time (Python has to look through every item one by one).

* **Dictionaries:** Finding a value by key takes $O(1)$ time (Python uses a **Hash Table** to jump straight to the data).

### **Q4: What is Dictionary Comprehension?**

**Answer:** Similar to List Comprehension, it's a one-liner to create a dictionary.
*Example:* `squared_dict = {x: x*x for x in range(5)}` results in `{0: 0, 1: 1, 2: 4, 3: 9, 4: 16}`.
