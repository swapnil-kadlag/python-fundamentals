

## 2. The Python Dictionary (`dict`)

A Dictionary is a collection of Key-Value pairs. It is optimized for retrieving data when you know the "name" (key) of the data rather than its position.

### 2.1 Essential Dictionary Methods

To interact with data efficiently, the following methods are utilized:

| Method | Syntax | Description |
| --- | --- | --- |
| **Access** | `.get(key, default)` | Retrieves value safely; returns `default` if key is missing. |
| **Inspection** | `.keys()` | Returns a view object of all keys in the dictionary. |
| **Inspection** | `.values()` | Returns a view object of all values in the dictionary. |
| **Iteration** | `.items()` | Returns a set of tuples representing `(key, value)` pairs. |
| **Modification** | `.update(obj)` | Merges another dictionary or iterable into the current one. |
| **Removal** | `.pop(key)` | Removes the specified key and returns the associated value. |
| **Cleanup** | `.clear()` | Removes all items from the dictionary. |

## 3. The Python Set (`set`)

A Set is an unordered collection of **unique** elements. It is the programmatic equivalent of a mathematical set.

### 3.1 Key Characteristics

* **No Duplicates:** Automatically filters out redundant data.
* **Unordered:** Items do not have a fixed position (no indexing).
* **Speed:** Highly optimized for checking if an item exists within the collection ($O(1)$ time complexity).

## 4. Practical Implementation (`dictionary_mastery.py`)

```
# Day 02: Advanced Dictionary and Set Implementation

# --- Dictionary Application ---
employee_data = {
    "id": 1024,
    "name": "Abhishek",
    "department": "AI/ML",
    "skills": ["Python", "Git"]
}

# Using .get() to avoid KeyError
salary = employee_data.get("salary", "Negotiable")

# Updating multiple values at once
employee_data.update({"location": "Pune", "department": "Gen AI"})

# Looping through Key-Value pairs
for key, value in employee_data.items():
    print(f"Field: {key} | Value: {value}")

# --- Set Application ---
# Fast duplicate removal
raw_logs = ["ERR_01", "ERR_02", "ERR_01", "ERR_03", "ERR_02"]
unique_errors = set(raw_logs)
print(f"Unique Error Codes: {unique_errors}")
```

## 5. Architectural Insights (Theory)

* **Hashing:** Dictionaries use a hash function to transform keys into addresses. This allows for constant-time $O(1)$ lookups, making them significantly faster than Lists for large-scale data retrieval.
* **Mutability:** Dictionaries and Sets are **mutable** (you can change them), but the **keys** of a dictionary must be **immutable** (like strings, numbers, or tuples).
